```python
import cv2
import matplotlib.pyplot as plt

# Ensure car_img is defined (moved from previous cell for robustness)
car_img = cv2.imread("/content/sample_data/car.jpg", cv2.IMREAD_COLOR)

# Check if image was loaded successfully
if car_img is None:
    print("Error: Could not load image. Please ensure '/car.jpg' exists and the path is correct.")
else:
    # Get the image dimensions (height, width)
    (h, w) = car_img.shape[:2]

    # Calculate the center of the image
    center = (w // 2, h // 2)

    # Define the rotation matrix
    # Here, we rotate by 90 degrees clockwise (positive angle), with no scaling
    M = cv2.getRotationMatrix2D(center, -90, 1.0)

    # Perform the rotation
    rotated_car_img = cv2.warpAffine(car_img, M, (w, h))

    # Convert the rotated image from BGR to RGB for correct display with Matplotlib
    rotated_car_img_rgb = cv2.cvtColor(rotated_car_img, cv2.COLOR_BGR2RGB)

    # Display the original and rotated images for comparison
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(car_img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rotated_car_img_rgb)
    plt.title('Rotated Image (-90 degrees)')
    plt.axis('off')

    plt.show()
```

```python
#Program 2
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("/content/sample_data/car.jpg")
assert img is not None, "Image not found!"

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows, cols = img.shape[:2]

plt.figure(figsize=(12, 10))

# Original Image
plt.subplot(3, 4, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

# Cropping
cropped = img[50:300, 100:400]
plt.subplot(3, 4, 2)
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis("off")

# Resizing
resized = cv2.resize(img, (300, 300))
plt.subplot(3, 4, 3)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis("off")

# Translation
M_trans = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(img, M_trans, (cols, rows))
plt.subplot(3, 4, 4)
plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.axis("off")

# Rotation
center = (cols // 2, rows // 2)
M_rot = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M_rot, (cols, rows))
plt.subplot(3, 4, 5)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.axis("off")

# Scaling
scaled = cv2.resize(img, None, fx=1.5, fy=1.5)
plt.subplot(3, 4, 6)
plt.imshow(cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB))
plt.title("Scaled Image")
plt.axis("off")

# Shearing
M_shear = np.float32([[1, 0.5, 0], [0, 1, 0]])
sheared = cv2.warpAffine(img, M_shear, (int(cols * 1.5), rows))
plt.subplot(3, 4, 7)
plt.imshow(cv2.cvtColor(sheared, cv2.COLOR_BGR2RGB))
plt.title("Sheared Image")
plt.axis("off")

# Reflection
flipped = cv2.flip(img, 1)
plt.subplot(3, 4, 8)
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title("Reflected Image")
plt.axis("off")

# Perspective Transformation
pts1 = np.float32([[50, 50], [cols - 50, 50], [50, rows - 50], [cols - 50, rows - 50]])
pts2 = np.float32([[10, 100], [cols - 100, 50], [100, rows - 10], [cols - 50, rows - 50]])

M_persp = cv2.getPerspectiveTransform(pts1, pts2)
perspective = cv2.warpPerspective(img, M_persp, (cols, rows))

plt.subplot(3, 4, 9)
plt.imshow(cv2.cvtColor(perspective, cv2.COLOR_BGR2RGB))
plt.title("Perspective Transform")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Program 1
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read images
cb_img = cv2.imread("/content/sample_data/chessboard.png", cv2.IMREAD_GRAYSCALE)
avocado_img = cv2.imread("/content/sample_data/avacado.jpg")

if cb_img is None or avocado_img is None:
    print("Image not found")
    exit()

# Image properties
print("Chessboard Shape:", cb_img.shape)
print("Chessboard Size:", cb_img.size)
print("Chessboard Data type:", cb_img.dtype)

# Conversions
avocado_rgb = cv2.cvtColor(avocado_img, cv2.COLOR_BGR2RGB)
avocado_gray = cv2.cvtColor(avocado_img, cv2.COLOR_BGR2GRAY)

# Split channels
b, g, r = cv2.split(avocado_img)

# Merge channels
merged_img = cv2.merge((b, g, r))
merged_rgb = cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB)

# HSV processing
hsv_img = cv2.cvtColor(avocado_img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img)
h_new = (h + 30) % 180
merged_hsv = cv2.merge((h_new, s, v))
final_img = cv2.cvtColor(merged_hsv, cv2.COLOR_HSV2RGB)

# Single-page display
fig, axs = plt.subplots(3, 3, figsize=(14, 12))
fig.suptitle("Color Image Processing Operations", fontsize=14)

axs[0, 0].imshow(cb_img, cmap="gray")
axs[0, 0].set_title("Grayscale Chessboard")
axs[0, 0].axis("off")

axs[0, 1].imshow(avocado_rgb)
axs[0, 1].set_title("Original Color Image")
axs[0, 1].axis("off")

axs[0, 2].imshow(avocado_gray, cmap="gray")
axs[0, 2].set_title("BGR to Grayscale")
axs[0, 2].axis("off")

axs[1, 0].imshow(b, cmap="gray")
axs[1, 0].set_title("Blue Channel")
axs[1, 0].axis("off")

axs[1, 1].imshow(g, cmap="gray")
axs[1, 1].set_title("Green Channel")
axs[1, 1].axis("off")

axs[1, 2].imshow(r, cmap="gray")
axs[1, 2].set_title("Red Channel")
axs[1, 2].axis("off")

axs[2, 0].imshow(merged_rgb)
axs[2, 0].set_title("Merged BGR Image")
axs[2, 0].axis("off")

axs[2, 1].imshow(final_img)
axs[2, 1].set_title("Hue Modified Image")
axs[2, 1].axis("off")

axs[2, 2].axis("off")  # Empty cell

plt.tight_layout()
plt.show()

# Save processed image
cv2.imwrite("/content/sample_data/avacado_saved.jpg", final_img)
```
